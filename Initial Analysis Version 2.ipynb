```python
# chosing songs from US only
song_rank_df.drop(song_rank_df[song_rank_df['Region'] != 'us'].index, inplace=True)
```

```python
# Confirming the regions are United States only
song_rank_df["Region"].unique()
```

```python
# to display the untruncated column of url
pd.set_option('display.max_colwidth', None)
song_rank_df["URL"]
```

```python
#url = "https://open.spotify.com/track/"
url_column = song_rank_df['URL']


regex = r"https:(.*?)\/track(.)"

subst = ""
url_column

# re.sub(pattern, repl, string, count=0, flags=0)
# You can manually specify the number of replacements by changing the 4th argument
result = []

for url_column in song_rank_df['URL']:
    result.append(re.sub(regex, subst, str(url_column),0, flags= re.MULTILINE))

# if result:
#     print (result)
```

```python
song_rank_df
```

```python
# Drop URL coulmn 

song_rank_df.drop("URL", axis=1, inplace=True)
```

```python
# Add the result column to the data frame

song_rank_df["new_URL"]= result
song_rank_df
```

```python
spotify_features_df.count()
```

```python
# Drop duplicates
spotify_features_df.drop_duplicates(subset="track_id", keep='first', inplace=True)
spotify_features_df
```

```python
#url = "https://open.spotify.com/track/"
artist_name_column = spotify_features_df['artist_name']


regex = r","

subst = " "
artist_name_column

# re.sub(pattern, repl, string, count=0, flags=0)
# You can manually specify the number of replacements by changing the 4th argument
result2 = []

for artist_name_column in spotify_features_df['artist_name']:
    result2.append(re.sub(regex, subst, str(artist_name_column),0, flags= re.MULTILINE))

# if result:
#     print (result)


```

```python
# Drop URL coulmn 

spotify_features_df.drop("artist_name", axis=1, inplace=True)
```

```python
# Add the result column to the data frame

spotify_features_df["new_artist_name"]= result2
spotify_features_df
```
